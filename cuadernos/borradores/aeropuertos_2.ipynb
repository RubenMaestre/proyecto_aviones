```python
# Definir los aeropuertos de interés
aeropuertos_de_interes = [
    "Dallas/Fort Worth International",
    "Ellison Onizuka Kona International at Keahole",
    "Guam International",
    "Francisco C. Ada Saipan International",
    "Pago Pago International",
    "Alexandria International",
    "Bill and Hillary Clinton National|Adams Field",
    "Montgomery Regional",
    "Ontario International",
    "Memphis International",
    "Stillwater Regional",
    "Birmingham-Shuttlesworth International",
    "Lovell Field|Chattanooga Metropolitan",
    "Tri-State/Milton J. Ferguson Field,"
    "Lafayette Regional Paul Fournet Field",
    "Gulfport-Biloxi International",
    "Albert J Ellis",
    "Pensacola International",
    "Huntsville International-Carl T Jones Field"
]

# Filtrar df_aeropuertos_unicos para obtener solo los aeropuertos de interés
df_interes = df_aeropuertos_unicos[df_aeropuertos_unicos['nombre_aeropuerto'].isin(aeropuertos_de_interes)]

# Crear un diccionario a partir del DataFrame filtrado
aeropuertos_dict = df_interes.set_index('nombre_aeropuerto')[['ciudad', 'estado', 'latitude', 'longitude', 'direccion']].to_dict(orient='index')
```

```python
# Función para buscar y reemplazar los datos basándose en el diccionario
def buscar_y_reemplazar(fila, columna_aeropuerto, es_origen=True):
    aeropuerto = fila[columna_aeropuerto]
    if aeropuerto in aeropuertos_dict:
        datos_aero = aeropuertos_dict[aeropuerto]
        prefijo = 'origen' if es_origen else 'destino'
        fila[f'ciudad_{prefijo}'] = datos_aero['ciudad']
        fila[f'estado_{prefijo}'] = datos_aero['estado']
        fila[f'latitude_{prefijo}'] = datos_aero['latitude']
        fila[f'longitude_{prefijo}'] = datos_aero['longitude']
        fila[f'direccion_{prefijo}'] = datos_aero['direccion']
    return fila

# Aplicar la función a cada fila para los aeropuertos de origen y destino
df_aviones = df_aviones.apply(buscar_y_reemplazar, columna_aeropuerto='aeropuerto_origen', es_origen=True, axis=1)
df_aviones = df_aviones.apply(buscar_y_reemplazar, columna_aeropuerto='aeropuerto_destino', es_origen=False, axis=1)
```

```python
# Definir la función que implementa la fórmula de Haversine (viene tal cual de chatgpt... parece que funciona)
# Definir la función que implementa la fórmula de Haversine
def haversine(lat1, lon1, lat2, lon2):
    # Convertir latitudes y longitudes de grados a radianes
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Diferencia de latitudes y longitudes
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Aplicar la fórmula de Haversine
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radio de la Tierra en millas
    R = 3958.8
    distance = R * c
    return distance

# Aplicar la función al DataFrame para calcular la distancia y crear una nueva columna
df_aviones['nueva_distancia_millas'] = df_aviones.apply(
    lambda row: haversine(
        row['latitude_origen'],
        row['longitude_origen'],
        row['latitude_destino'],
        row['longitude_destino']
    ),
    axis=1
)
```

```python
# Paso 1: Filtrar valores nulos y reducir a valores únicos
df_filtrado = df_aviones.dropna(subset=['distancia_millas', 'nueva_distancia_millas'])
df_unicos = df_filtrado.drop_duplicates(subset=['distancia_millas', 'nueva_distancia_millas'])

# Paso 2: Calcular la desviación porcentual
df_unicos['desviacion_pct'] = np.abs(df_unicos['nueva_distancia_millas'] - df_unicos['distancia_millas']) / df_unicos['distancia_millas'] * 100

# Paso 3: Categorizar las desviaciones
rangos = [100, 80, 50, 35, 15, 10, 5, 0]
conteos = {}

for i in range(len(rangos) - 1):
    min_rango = rangos[i + 1]
    max_rango = rangos[i]
    conteo = ((df_unicos['desviacion_pct'] > min_rango) & (df_unicos['desviacion_pct'] <= max_rango)).sum()
    conteos[f'Más allá del {min_rango}% y hasta el {max_rango}%'] = conteo

# Cuantos están en menos del 5%
conteos['Menos del 5%'] = (df_unicos['desviacion_pct'] <= 5).sum()

# Imprimir los resultados
for rango, conteo in conteos.items():
    print(f"{rango}: {conteo}")
```
